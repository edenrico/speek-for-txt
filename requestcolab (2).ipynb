```python
import whisper
from pyannote.audio import Pipeline, Model
from pydub import AudioSegment, effects
import os
from google.colab import files
import concurrent.futures

# Remover conversão para PCM, já que o Whisper suporta MP3
def preprocess_audio(audio_file):
    audio = AudioSegment.from_file(audio_file)
    audio = effects.normalize(audio)
    processed_file = "processed_" + os.path.basename(audio_file)
    audio.export(processed_file, format="mp3")
    return processed_file, audio  # Retornar também o objeto audio

# Função para realizar diarização
def diarize_audio(audio_file, token):
    diarization_model = Model.from_pretrained("pyannote/segmentation", use_auth_token=token)
    diarization_pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization", use_auth_token=token)
    diarization_result = diarization_pipeline(audio_file)
    return diarization_result

# Função para transcrever um segmento
def transcribe_segment(segment_file):
    return modelo.transcribe(segment_file)['text']

# Carregar e pré-processar o áudio
audio_file = '/content/PABLO_MARCAL_CUTUCA_CADA_CANDIDATO_a_PREFEITURA_de_SAO_PAULO_e_DIZ_PORQUE_NAO_APOIAR_ELES__[_YouConvert.net_].mp3'
processed_audio_file, audio = preprocess_audio(audio_file)

# Carregar o modelo Whisper utilizando a GPU
print("Carregando o modelo Whisper medium com GPU...")
modelo = whisper.load_model('medium', device='cuda')  # Utilizando a GPU (cuda)

# token de autenticação do Hugging Face
hf_token = "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"

# Realizar diarização
print("Realizando diarização...")
diarization_result = diarize_audio(processed_audio_file, hf_token)

# Transcrever o áudio com separação de falantes de forma paralela
transcript = ""
with concurrent.futures.ThreadPoolExecutor() as executor:
    future_to_speaker = {}
    for turn, _, speaker in diarization_result.itertracks(yield_label=True):
        # Extrair o segmento de áudio correspondente ao falante
        segment = audio[turn.start * 1000:turn.end * 1000]  # converter para milissegundos
        segment_file = f"segment_{speaker}.mp3"
        segment.export(segment_file, format="mp3")  # Salvar segmento temporário

        # Submeter a transcrição do segmento para processamento paralelo
        future = executor.submit(transcribe_segment, segment_file)
        future_to_speaker[future] = speaker

    # Coletar os resultados conforme forem concluídos
    for future in concurrent.futures.as_completed(future_to_speaker):
        speaker = future_to_speaker[future]
        try:
            transcript += f"Falante {speaker}: {future.result()}\n"
        except Exception as e:
            print(f"Erro ao transcrever: {e}")

# Salvar a transcrição em um arquivo de texto
output_file = 'transcricao_diarizada.txt'
with open(output_file, 'w') as file:
    file.write(transcript)

# Baixar o arquivo de texto transcrito
files.download(output_file)

print(f"Transcrição diarizada salva em {output_file}")
```
